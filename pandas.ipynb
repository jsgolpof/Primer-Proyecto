```python
import pandas as pd
import numpy as np

chipo = pd.read_csv('chipotle.tsv', sep = '\t')

chipo.head()
```

```python
# chipo.dropna() # preview de borra nulos
chipo = chipo.dropna() # borra nulos
chipo
```

```python
chipo.info() # info fichero
```

```python
orders = chipo.order_id.value_counts().count() # cuenta distintos
orders # Muestra lo almacenado
```

```python
dollarizer = lambda x: float(x[1:-1]) # funcion para eliminar símbolo a la derecha
chipo.item_price = chipo.item_price.apply(dollarizer) 
chipo
```

```python
chipo["revenue"] = chipo["quantity"] * chipo["item_price"]
chipo
```
